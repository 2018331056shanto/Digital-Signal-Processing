# Determine and plot the Fourier series of DT periodic signal.
###**Reg :2018331056**

####In digital signal processing, the term Discrete Fourier series (DFS) is any periodic discrete-time signal comprising harmonically-related (i.e. Fourier) discrete real sinusoids or discrete complex exponentials, combined by a weighted summation


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy.ma.core import sin
time=np.arange(0,2*np.pi,0.01)*(1/100)
F=60

signal=50*np.sin(2*np.pi*F*time)

plt.figure(figsize=(50,15))
plt.subplot(221)
plt.plot(time,signal)
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("CT Periodic Signal")
plt.show()



n=np.arange(60)
f=.05

dt=f/F

Dtsignal=np.sin(2*np.pi*n*F*dt)*50
# print(Dtsignal)
plt.figure(figsize=(50,15))
plt.subplot(222)
plt.stem(n,Dtsignal)
plt.grid()
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.title("DT Periodic Signal")
plt.show()


```
