## Convolution

####Convolution is a mathematical operation used to express the relation between input and output of an LTI system.

Screenshot from 2022-08-09 16-31-44.png

####The process of computing the convolution between x(k) and h(k) involves the following four steps.

 **Folding**. Fold h(k) about k = 0 to obtain h(−k).
-----
 **Shifting**. Shift h(−k) by n0 to the right (left) if n0 is positive (negative), to obtain h(n0 − k).
 ---------
 **Multiplication**. Multiply x(k) by h(n0 − k) to obtain the product sequence vn0 (k) ≡ x(k)h(n0 − k).
-------- 
 **Summation**. Sum all the values of the product sequence vn0 (k) to obtain the value of the output at time n = n0 .
-----------

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Input Signal && Impulse Response

```python
x=[1,2,3,1] #input signal
h=[1,2,1,-1] #impulse response
# plt.plot(x,h)
n=np.arange(-10,10)
inputSamples=np.zeros(n.shape)
idx=0;
for i in range(10,10+len(x),1):
  inputSamples[i]=x[idx]
  idx+=1
# print(inputSamples)

# print(input)
# plt.plot(input)
plt.figure(figsize=(20,10))
plt.subplot(221)
plt.title("Input Signal x(n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,inputSamples)
plt.show()

n=np.arange(-10,10)
impulseSamples=np.zeros(n.shape)
idx=0;
for i in range(9,9+len(h),1):
  impulseSamples[i]=h[idx]
  idx+=1
# print(impulseSamples)

# print(input)
# plt.plot(input)
plt.figure(figsize=(20,10))
plt.subplot(221)
plt.title("Impulse Response h(n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,impulseSamples)
plt.show()




```

## Floding the Impulse Response

```python
#folding
temp=h[::-1]
# print(temp)
foldImpulseResponse=np.zeros(n.shape)
idx=0
for i in range(9,9+len(h),1):
  foldImpulseResponse[i]=temp[idx]
  idx+=1
plt.figure(figsize=(20,10))
plt.subplot(221)
plt.title("Impulse Response h(-n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,foldImpulseResponse)
plt.show()


```

## Multiplication and Summation

```python
#multiplication

impulse=np.zeros(n.shape)
multilyWithImpulse=np.zeros(n.shape)
impulse=foldImpulseResponse
ans=[]
# print(foldImpulseResponse)
# print(impulse)
for i in range(-2,5,1):
  # print(i)
  impulse=np.roll(foldImpulseResponse,i)
  # print(impulse)
  plt.figure(figsize=(20,10))
  plt.subplot(221)
  st="impulse response h(-n ) "+str(i);
  # print(st)
  plt.title(st)
  plt.xlabel("Samples")
  plt.ylabel("Amplitude")
  plt.xticks(n)
  plt.stem(n,impulse)
  plt.show()

  multilyWithImpulse=np.multiply(impulse,inputSamples)
  #Summation
  ans.append(np.sum(multilyWithImpulse))
  plt.figure(figsize=(20,10))
  plt.subplot(221)
  st="Multiplication h(-n ) "+str(i);
  # print(st)
  plt.title(st)
  plt.xlabel("Samples")
  plt.ylabel("Amplitude")
  plt.xticks(n)
  plt.stem(n,multilyWithImpulse)
  plt.show()


print("Convolution of Two signal x(n) and h(n) :",ans)



  


```
