# Compute Discrete Fourier Transform (DFT) of a signal using DFT equation.

### **Reg :2018331056**

###As the name implies, the Discrete Fourier Transform (DFT) is purely discrete: discrete-time data sets are converted into a discrete-frequency representation

Screenshot from 2022-08-10 12-10-17.png

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def compute_DFT(x_n):

  X_k=np.zeros(len(x_n),dtype=np.complex)
  for k in range(0,len(x_n),1):
    temp=0+0j
    for n in range(len(x_n)):
      temp+=(x_n[n]*np.exp((-1)*(1j*2*np.pi*k*n)/len(x_n)))
    
    # print(temp)
      X_k[k]=temp
  return X_k


x_n=np.array([0,1,2,3]) #input signal
# print(type(x_n))
X_k=compute_DFT(x_n)


magnitude=np.absolute(X_k)
phase=np.angle(X_k)

plt.figure(figsize=(30,10))
plt.subplot(221)
plt.title("Input Sample x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(x_n)
plt.stem(x_n)
plt.show()

# print(magnitude)
# print(phase)

plt.figure(figsize=(30,10))
plt.subplot(222)
plt.title("DFT  X(k) Magnitude")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(x_n)
plt.stem(magnitude)
plt.show()


plt.figure(figsize=(30,10))
plt.subplot(223)
plt.title("DFT  X(k) Phase")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(x_n)
plt.stem(phase)
plt.show()
```
