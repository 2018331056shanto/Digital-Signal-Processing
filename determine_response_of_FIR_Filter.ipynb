# By means of the DFT and IDFT, determine the response of the FIR filter with impulse response h(n) to the input sequence x(n)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Compute DFT of Input Signal x(n)

```python
def compute_DFT(x_n):
  N=8
  X_k=np.zeros(N,dtype=np.complex)
  for k in range(0,N,1):
    temp=0+0j
    for n in range(len(x_n)):
      temp+=(x_n[n]*np.exp((-1)*(1j*2*np.pi*k*n)/N))
    
    # print(temp)
      X_k[k]=temp
  return X_k


x=[1,2,2,1]

x_n=np.array(x) #input signal
# print(type(x_n))
X_k=compute_DFT(x_n)
# print(X_k)

magnitude=np.absolute(X_k)
phase=np.angle(X_k)
n=np.arange(8)

plt.figure(figsize=(30,10))
plt.subplot(221)
plt.title("Input Sample x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(x_n)
plt.stem(x_n)
plt.show()

# print(magnitude)
# print(phase)

plt.figure(figsize=(30,10))
plt.subplot(222)
plt.title("DFT  X(k)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(X_k)
plt.show()

plt.figure(figsize=(30,10))
plt.subplot(222)
plt.title("DFT  X(k) Magnitude")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(magnitude)
plt.show()


plt.figure(figsize=(30,10))
plt.subplot(223)
plt.title("DFT  X(k) Phase")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(n)
plt.stem(phase)
plt.show()

print("DFT of x(n) :",X_k)
```

## Compute DFT of Impulse Response h(n)

```python
def compute_DFT(h_n):
  N=8
  H_k=np.zeros(N,dtype=np.complex)
  for k in range(0,N,1):
    temp=0+0j
    for n in range(len(h_n)):
      temp+=(h_n[n]*np.exp((-1)*(1j*2*np.pi*k*n)/N))
    
    # print(temp)
      H_k[k]=temp
  return H_k


h=[1,2,3]

h_n=np.array(h) #input signal
# print(type(x_n))
H_k=compute_DFT(h_n)
# print(H_k)

magnitude=np.absolute(H_k)
phase=np.angle(H_k)

plt.figure(figsize=(30,10))
plt.subplot(221)
plt.title("Input Sample x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(h_n)
plt.stem(h_n)
plt.show()

# print(magnitude)
# print(phase)

plt.figure(figsize=(30,10))
plt.subplot(223)
plt.title("DFT  H(k)")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(n)
plt.stem(H_k)
plt.show()

plt.figure(figsize=(30,10))
plt.subplot(222)
plt.title("DFT  H(k) Magnitude")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(magnitude)
plt.show()


plt.figure(figsize=(30,10))
plt.subplot(223)
plt.title("DFT  H(k) Phase")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(n)
plt.stem(phase)
plt.show()

print("DFT of h(n) :",H_k)
```

## Multiply H(K) with X(k)

```python
Y_k=np.multiply(X_k,H_k)
plt.figure(figsize=(30,10))
plt.subplot(223)
plt.title("DFT  Y(k) ")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(n)
plt.stem(Y_k)
plt.show()
print(Y_k)
```

## IDFT of Y(K)

```python
def compute_IDFT(Y_k): 
  y_n=np.zeros(len(Y_k))

  for n in range(len(Y_k)):
    temp=0+0j
    for k in range(len(Y_k)):
      temp+=(Y_k[k]*np.exp((2*np.pi*1j*n*k/len(Y_k))))
    # print(temp)
    y_n[n]=temp/len(Y_k)
  return y_n
 

y_n=compute_IDFT(Y_k)

plt.figure(figsize=(30,15))
plt.subplot(223)
plt.title("x(n)=IDFT X(k)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(y_n)
plt.stem(y_n)
plt.show()
print("IDFT by equation :",y_n)
# print(X_n)
```
