### The discrete Fourier transform is an invertible, linear transformation. with denoting the set of complex numbers. Its inverse is known as Inverse Discrete Fourier Transform (IDFT). In other words, for any , an N-dimensional complex vector has a DFT and an IDFT which are in turn. -dimensional complex vectors.

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

## Computing DFT

```python
def compute_DFT(x_n):

  X_k=np.zeros(len(x_n),dtype=np.complex)
  for k in range(0,len(x_n),1):
    temp=0+0j
    for n in range(len(x_n)):
      temp+=(x_n[n]*np.exp((-1)*(1j*2*np.pi*k*n)/len(x_n)))
    
    # print(temp)
      X_k[k]=temp
  return X_k


x_n=np.array([0,1,2,3]) #input signal
# print(type(x_n))
X_k=compute_DFT(x_n)
# print(X_k)

magnitude=np.absolute(X_k)
phase=np.angle(X_k)

plt.figure(figsize=(30,15))
plt.subplot(221)
plt.title("Input Sample x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(x_n)
plt.stem(x_n)
plt.show()

# print(magnitude)
# print(phase)

plt.figure(figsize=(30,15))
plt.subplot(222)
plt.title("DFT  X(k) Magnitude")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(x_n)
plt.stem(magnitude)
plt.show()


plt.figure(figsize=(30,15))
plt.subplot(223)
plt.title("DFT  X(k) Phase")
plt.xlabel("Sample")
plt.ylabel("Ampliitude")
plt.xticks(x_n)
plt.stem(phase)
plt.show()

print("Computin DFT by Equation :",X_k)
```

## Computing IDFT Using Equation

```python
def compute_IDFT(X_k): 
  X_n=np.zeros(len(X_k))

  for n in range(len(X_k)):
    temp=0+0j
    for k in range(len(X_k)):
      temp+=(X_k[k]*np.exp((2*np.pi*1j*n*k/len(X_k))))
    # print(temp)
    X_n[n]=temp/len(X_k)
  return X_n
 

X_n=compute_IDFT(X_k)

plt.figure(figsize=(30,15))
plt.subplot(223)
plt.title("x(n)=IDFT X(k)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(X_n)
plt.stem(X_n)
plt.show()
print("IDFT by equation :",X_n)
# print(X_n)
```

## Computing IDFT Using python library numpy.fft.ifft

```python
idft_py_library=np.zeros(len(X_k))
idft_py_library=np.fft.ifft(X_k)


plt.figure(figsize=(30,15))
plt.subplot(223)
plt.title("x(n)=IDFT X(k)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.stem(idft_py_library)
plt.show()

print("Idft by python library :",idft_py_library)
```
