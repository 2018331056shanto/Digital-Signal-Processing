## Correlation
Correlation is a simple mathematical operation to compare two signals. Correlation is also a convolution operation between two signals. But there is a basic difference. Correlation of two signals is the convolution between one signal with the functional inverse version of the other signal.

The special case where y(n) = x(n), we have the **autocorrelation** of

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Input Signal

```python
n=np.arange(-20,20)
x=[2,-1,3,7,1,2,-3]
inputSample=np.zeros(n.shape)

idx=0
for i in range(16,16+len(x),1):
  inputSample[i]=x[idx]
  idx+=1

plt.figure(figsize=(30,15))
plt.subplot(221)
plt.title("Input Signal x(n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.yticks(n)
plt.xticks(n)
plt.stem(n,inputSample)
plt.show()
```

## Floding the Impulse Response

```python
#folding
temp=x[::-1]
# print(temp)
foldImpulseResponse=np.zeros(n.shape)
idx=0
for i in range(16,16+len(x),1):
  foldImpulseResponse[i]=temp[idx]
  idx+=1
foldImpulseResponse=np.roll(foldImpulseResponse,2)
plt.figure(figsize=(30,15))
plt.subplot(221)
plt.title("Folded Signal x(-n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.yticks(n)
plt.xticks(n)
plt.stem(n,foldImpulseResponse)
plt.show()
```

## Multiplication and Summation

```python
impulse=np.zeros(n.shape)
multilyWithImpulse=np.zeros(n.shape)
impulse=foldImpulseResponse
ans=[]
# print(foldImpulseResponse)
# print(impulse)
for i in range(-8,5,1):
  # print(i)
  impulse=np.roll(foldImpulseResponse,i)
  # print(impulse)
  plt.figure(figsize=(30,15))
  plt.subplot(221)
  st="impulse response h(-n ) "+str(i);
  # print(st)
  plt.title(st)
  plt.xlabel("Samples")
  plt.ylabel("Amplitude")
  plt.xticks(n)
  plt.stem(n,impulse)
  plt.show()

  multilyWithImpulse=np.multiply(impulse,inputSample)
  #Summation
  ans.append(np.sum(multilyWithImpulse))
  plt.figure(figsize=(30,15))
  plt.subplot(221)
  st="Multiplication h(-n ) "+str(i);
  # print(st)
  plt.title(st)
  plt.xlabel("Samples")
  plt.ylabel("Amplitude")
  plt.xticks(n)
  plt.stem(n,multilyWithImpulse)
  plt.show()


print("Auto Correlation  of Two signal x(n) and x(-n) :",ans)
```

## Final Output Signal

```python
finalAns=np.zeros(n.shape)
m=np.arange(-40,40)
idx=0
for i in range(13,13+len(ans),1):
  finalAns[i]=ans[idx]
  idx+=1

plt.figure(figsize=(30,40))
plt.subplot(221)
st="Cross Correlation";
# print(st)
plt.title(st)
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.yticks(m)
plt.xticks(n)
plt.stem(n,finalAns)
plt.show()
```
