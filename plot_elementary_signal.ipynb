```python
import matplotlib.pyplot as plt
import numpy as np
import math
import timeit
```

```python
#unit impulse response
left=-10
right=10
#np arrange for evenly spaced values within a given interval
n=np.arange(left,right+1)
#np zeros return a new array of given shape and type filled with zeros
#np array have an attribute "shape" that returns tuple with each index having the number of corresponding elements
# plt.plot(n.shape)
response=np.zeros(n.shape)
# plt.plot(response)
#type(response)
#np where Return elements chosen from x or y depending on condition
response[np.where(n==0)]=1
#plt figure Create a new figure, or activate an existing figure
plt.figure(figsize=(10,6))
#plt subplot Subplots mean groups of axes that can exist in a single matplotlib figure.
plt.subplot(111)
plt.title("Unit Impulse Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
#plt xticks Get or set the current tick locations and labels of the x-axis. Pass no arguments to return the current values without modifying 
plt.xticks(n)
#plt stem A stem plot draws lines perpendicular to a baseline at each location locs from the baseline to heads, and places a marker there
plt.stem(n,response)
plt.show()


```

```python
#ramp
left=-10
right=10
n=np.arange(left,right+1)
unit_ramp=np.array([i if i>=0 else 0 for i in n])
# plt.plot(unit_ramp)
plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Ramp Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,unit_ramp)
plt.show()
```

```python
#Exponential
left=-10
right=10
n=np.arange(left,right+1)
a1 = 0.9    # 0 < a < 1
a2 = 1.1    # a > 1
a3 = -0.9   # -1 < a < 0
a4 = -1.1   # a < -1

x1 = a1**n
x2 = a2**n
x3 = a3**n
x4 = a4**n

plt.figure(figsize=(20,12))
plt.subplot(221)
plt.title("Exponential Signal 0<a<1")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,x1)


plt.figure(figsize=(20,12))
plt.subplot(222)
plt.title("Exponential Signal a>1")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,x2)

plt.figure(figsize=(20,12))
plt.subplot(223)
plt.title("Exponential Signal -1<a<0")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,x3)

plt.figure(figsize=(20,12))
plt.subplot(224)
plt.title("Exponential Signal a<-1")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,x4)


plt.show()





```

```python
#impulse response
# print(n)
left=-10
right=10
n=np.arange(left,right+1)

plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
plt.title("Exponential Signal a>1")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,x2)

response=np.zeros(n.shape)
response[np.where(n==0)]=1

# print(x2)
plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Impulse Signal delta(n)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,response)

# print(response)
response1=np.roll(response,3)
plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Impulse Signal delta(n-3)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,response1)

# print(response1)

ans=np.multiply(response1,x2)
# print(ans)
plt.figure(figsize=(12,6))
plt.subplot(111)
plt.title("Unit Impulse Signal x(n)*delta(n-3) ")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n,ans)
plt.show()
```
